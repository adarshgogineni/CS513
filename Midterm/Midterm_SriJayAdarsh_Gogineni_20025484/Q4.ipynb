```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from BP.csv
df = pd.read_csv('BP.csv')

# View the first few rows of the DataFrame to ensure it's loaded correctly
df.head()
```

```python
# Create a box plot for 'Age'
plt.figure(figsize=(10, 5))
sns.boxplot(x='Age', data=df)
plt.title('Box Plot of Age')
plt.show()

# Create a box plot for 'Height'
plt.figure(figsize=(10, 5))
sns.boxplot(x='Height', data=df)
plt.title('Box Plot of Height')
plt.show()

# Create a box plot for 'Weight'
plt.figure(figsize=(10, 5))
sns.boxplot(x='Weight', data=df)
plt.title('Box Plot of Weight')
plt.show()

# Create pair plots for 'Age', 'Height', and 'Weight'
sns.pairplot(df[['Age', 'Height', 'Weight']])
plt.suptitle('Pair Plots of Age, Height, and Weight', y=1.02)
plt.show()
```
