```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
data = pd.read_csv('BP.CSV')
```

```python
# Define the feature columns (excluding 'BP_Status')
features = data[['Somker', 'Gender', 'Age', 'Height', 'Weight']]

# Encode categorical variables (Smoker and Gender) using one-hot encoding
features = pd.get_dummies(features, columns=['Somker', 'Gender'], drop_first=True)

# Define the target variable ('BP_Status')
target = data['BP_Status']
```

```python
#TRAIN MODEL
# Split the data into training (70%) and test (30%) sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Initialize a list to store results for different k values
results = []

# Define a range of k values to test
k_values = [3, 5, 7, 9, 11, 13, 15]

# Iterate through different k values
for k in k_values:
    # Initialize the KNN classifier with the current k value
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    
    # Normalize the features using StandardScaler
    scaler = StandardScaler()
    X_train_normalized = scaler.fit_transform(X_train)
    X_test_normalized = scaler.transform(X_test)
    
    # Fit the model to the normalized training data
    knn_classifier.fit(X_train_normalized, y_train)
    
    # Make predictions on the normalized test data
    y_pred = knn_classifier.predict(X_test_normalized)
    
    # Calculate model metrics
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    
    # Store the results for this k value
    results.append({'k': k, 'accuracy': accuracy, 'f1': f1, 'recall': recall})

# Print the results for different k values
for result in results:
    print(f"K = {result['k']}:")
    print(f"Accuracy: {result['accuracy']:.2f}")
    print(f"F1 Score: {result['f1']:.2f}")
    print(f"Recall Score: {result['recall']:.2f}")
    print()
    # Display a classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
```
