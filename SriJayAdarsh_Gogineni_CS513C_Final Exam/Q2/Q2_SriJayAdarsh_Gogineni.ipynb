```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('NYNJ_zipcode_population.csv')  # Replace with your file path
```

```python
# Select the 'Low_income' percent and 'Total_Pop' columns for clustering
data_to_cluster = df[['Low_income', 'Total_Pop']]
```

```python
df.head()
```

```python
data_to_cluster.head()
```

```python
# Perform hierarchical clustering using the single linkage method
Z = linkage(data_to_cluster, method='single')
```

```python
# Plot the dendrogram to visualize the clusters
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
dendrogram(Z, labels=df['zipcode'].values)  
plt.ylabel('Distance')
plt.show()
```

```python
# obtain the specific members of each cluster, using the 'fcluster' function
#Set clusters to 3
from scipy.cluster.hierarchy import fcluster
desired_clusters = 3
clusters = fcluster(Z, t=desired_clusters, criterion='maxclust')
```

```python
# Add the cluster labels to dataframe
df['Cluster_Label'] = clusters
```

```python
# see which zip codes are in which clusters
cluster_members = df.groupby('Cluster_Label')['zipcode'].apply(list) 
```

```python
for label, members in cluster_members.items():
    print(f"Cluster {label}: {members}")
```

### Cluster 1 contains a large number of zip codes.
### Cluster 2 and Cluster 3 each contain a few zip codes, but not just one.
### In this context, we could consider the zip codes in Clusters 2 and 3 as more unusual compared to those in Cluster 1 simply because they are grouped into smaller clusters, which suggests that their combination of "Low_income" percent and "Total_Pop" is less common in the dataset.

### However, to truly determine if there is an "unusual" zip code, we'd typically look for a cluster with only one member or examine the data to find any zip codes that are significantly different from others in their cluster. Since none of the clusters contain just one zip code, there isn't a single zip code that can be definitively labeled as unusual based on the provided information.
